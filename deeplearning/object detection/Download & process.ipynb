# Download and process images

```python
import os
from googledriver import download_folder
from PIL import Image
from pillow_heif import register_heif_opener
```

```python
url = "https://drive.google.com/drive/folders/1gBEBk37-45OzbbCctfWkm-vnKJ74XT4D"
output_path = "images" 

download_folder(url, output_path)
```

```python
register_heif_opener()

def convert_and_reduce_quality(heic_path, jpeg_path, quality_percentage=60):
    """
    Converts a single HEIC file to JPEG and reduces the quality/size.
    """
    try:
        img = Image.open(heic_path)
        if img.mode != 'RGB':
            img = img.convert('RGB')

        img.save(jpeg_path, format='JPEG', quality=quality_percentage)
        
        print(f"Successfully converted and saved to {jpeg_path} with quality: {quality_percentage}%")

    except Exception as e:
        print(f"An error occurred: {e}")

def get_image_list():
    # This function gets only the images in subdirectories, ignoring the store directory
    directory_path = './images/'
    all_subdir = [f"{directory_path}/{directory}/{os.listdir(directory_path + '/' + directory)} for directory in os.listdir(directory_path) if directory != 'store' and directory != 'file" and "." not in directory]
    print(all_subdir[0])

        
    return all_subdir
#    return final_list [f"{entry}/{image}" for entry in all_subdir for image in entry if image.lower().endswith(".heic") ]
```

```python
try:
    os.mkdir(f"./data/")
except:
    pass # donot
images = get_image_list()
total_images = len(images)
for i, image in enumerate(images):
    convert_and_reduce_quality(f"./images/{image}", f"./data/{image[:-5]}.jpg", 10)
    percentage = ((i + 1) * 100 / total_images)
    print(f"Percentage: {round(percentage, 2)} %")
```
